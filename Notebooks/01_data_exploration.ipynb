## Dataset Overview

The dataset contains **11,585 observations and 34 variables** collected from a **batch penicillin fermentation process** in an industrial-style bioreactor. No missing values are present. The variables consist of **23 floating-point measurements and 11 integer variables**.

Each row represents a **time point within a batch**, identified by `batch_id`. The variables describe operating conditions, biological state variables, and measured process outputs.

### Main Variable Categories

**1. Process Operating Conditions (Inputs / Control Variables)**
These describe how the reactor is operated and will later serve as key optimization variables.

* Aeration rate (L/h)
* Agitator RPM
* Sugar feed rate (L/h)
* Acid flow rate (L/h)
* Base flow rate (L/h)
* Heating/cooling water flow rate (L/h)
* Heating water flow rate (L/h)
* Water for injection/dilution (L/h)
* Air head pressure (bar)

**2. Bioprocess State Variables**
These describe the internal biological and chemical state of the fermentation.

* Substrate concentration (g/L)
* Dissolved oxygen concentration (mg/L)
* pH
* Temperature (K)
* Vessel volume (L)
* Vessel weight (kg)
* Viscosity (centipoise)

**3. Production and Biological Performance Variables**
These variables describe microbial activity and product formation.

* Penicillin concentration (g/L)
* Biomass concentration (offline, g/L)
* Oxygen uptake rate (g/min)
* Carbon evolution rate (g/h)
* CO₂ percentage in off-gas (%)
* O₂ percentage in off-gas (%)

**4. Feed and Chemical Additions**

* PAA flow (L/h)
* PAA concentration (g/L)
* Oil flow (L/h)
* Ammonia concentration (g/L)
* Ammonia shots (kg)

**5. Operational / Monitoring Variables**

* Generated heat (kJ)
* Dumped broth flow (L/h)
* Fault reference
* Control reference (recipe driven vs operator controlled)

### Units

Most variables use **standard bioprocess engineering units**, including:

* **Flow rates:** L/h
* **Concentrations:** g/L or mg/L
* **Pressure:** bar
* **Temperature:** Kelvin (K)
* **Gas composition:** %
* **Biological rates:** g/min or g/h
* **Viscosity:** centipoise

### Relevance to the Project

This dataset provides both **process control variables (RPM, aeration, feed rates)** and **biological response variables (biomass, penicillin concentration)**.

These features will be used to:

* **Model fermentation dynamics** over time
* **Predict biomass and penicillin concentration trajectories**
* **Identify optimal operating conditions** that maximize final penicillin yield
* **Simulate alternative operating strategies** for process optimization

The presence of detailed operational variables and gas-exchange measurements (OUR, CER, off-gas composition) makes the dataset suitable for **feature engineering, dynamic modeling, and future integration with mechanistic bioprocess models or digital twins**.
