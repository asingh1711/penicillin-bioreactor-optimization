## Dataset Overview

The dataset contains **11,585 observations and 34 variables** collected from a **batch penicillin fermentation process** in an industrial-style bioreactor. No missing values are present. The variables consist of **23 floating-point measurements and 11 integer variables**.

Each row represents a **time point within a batch**, identified by `batch_id`. The variables describe operating conditions, biological state variables, and measured process outputs.

### Main Variable Categories

**1. Process Operating Conditions (Inputs / Control Variables)**
These describe how the reactor is operated and will later serve as key optimization variables.

* Aeration rate (L/h)
* Agitator RPM
* Sugar feed rate (L/h)
* Acid flow rate (L/h)
* Base flow rate (L/h)
* Heating/cooling water flow rate (L/h)
* Heating water flow rate (L/h)
* Water for injection/dilution (L/h)
* Air head pressure (bar)

**2. Bioprocess State Variables**
These describe the internal biological and chemical state of the fermentation.

* Substrate concentration (g/L)
* Dissolved oxygen concentration (mg/L)
* pH
* Temperature (K)
* Vessel volume (L)
* Vessel weight (kg)
* Viscosity (centipoise)

**3. Production and Biological Performance Variables**
These variables describe microbial activity and product formation.

* Penicillin concentration (g/L)
* Biomass concentration (offline, g/L)
* Oxygen uptake rate (g/min)
* Carbon evolution rate (g/h)
* CO₂ percentage in off-gas (%)
* O₂ percentage in off-gas (%)

**4. Feed and Chemical Additions**

* PAA flow (L/h)
* PAA concentration (g/L)
* Oil flow (L/h)
* Ammonia concentration (g/L)
* Ammonia shots (kg)

**5. Operational / Monitoring Variables**

* Generated heat (kJ)
* Dumped broth flow (L/h)
* Fault reference
* Control reference (recipe driven vs operator controlled)

### Units

Most variables use **standard bioprocess engineering units**, including:

* **Flow rates:** L/h
* **Concentrations:** g/L or mg/L
* **Pressure:** bar
* **Temperature:** Kelvin (K)
* **Gas composition:** %
* **Biological rates:** g/min or g/h
* **Viscosity:** centipoise

### Relevance to the Project

This dataset provides both **process control variables (RPM, aeration, feed rates)** and **biological response variables (biomass, penicillin concentration)**.

These features will be used to:

* **Model fermentation dynamics** over time
* **Predict biomass and penicillin concentration trajectories**
* **Identify optimal operating conditions** that maximize final penicillin yield
* **Simulate alternative operating strategies** for process optimization

The presence of detailed operational variables and gas-exchange measurements (OUR, CER, off-gas composition) makes the dataset suitable for **feature engineering, dynamic modeling, and future integration with mechanistic bioprocess models or digital twins**.


```python
time_col = "Time (h)"

variables = {
    "biomass": "Offline Biomass concentratio(X_offline:X(g L^{-1}))",
    "penicillin": "Penicillin concentration(P:g/L)",
    "RPM": "Agitator RPM(RPM:RPM)",
    "feed_rate": "Sugar feed rate(Fs:L/h)",
    "O2": "Oxygen in percent in off-gas(O2:O2  (%))",
    "CO2": "carbon dioxide percent in off-gas(CO2outgas:%)"
}

for name, col in variables.items():
    plt.figure(figsize=(6,4))
    sns.lineplot(data=df, x=time_col, y=col)
    plt.title(name)
    plt.xlabel("Time (h)")
    plt.ylabel(name)
    plt.show()
```

Biomass increases rapidly from ~0 to about 25 g/L by ~150 h, indicating active microbial growth during the early fermentation phase. After ~150–170 h, biomass plateaus and gradually declines to around 18–20 g/L, suggesting nutrient limitation or the transition to secondary metabolism. Penicillin production begins slowly but rises sharply after ~40–50 h, reaching a peak of about 27–28 g/L around 200–220 h. This indicates that penicillin synthesis occurs mainly after the rapid biomass growth phase.

The feed rate increases stepwise early in the process (peaking ~150 L/h) to support biomass growth, then stabilizes near 80–90 L/h. O₂ concentration drops initially (≈0.205 to ≈0.182), reflecting high oxygen consumption during growth, and gradually recovers as growth slows. CO₂ rises from near 0 to ~2% by ~90 h, showing increased metabolic activity, then decreases after ~150 h.

A clear pattern appears: biomass growth correlates with increased CO₂ and decreased O₂, indicating active respiration. Penicillin production increases after biomass stabilizes, suggesting it is a secondary metabolite produced during the later fermentation phase.

```python
data = df.copy()

# Remove non-informative columns
cols_to_drop = [
    'batch_id',
    'Fault reference(Fault_ref:Fault ref)',
    '0 - Recipe driven 1 - Operator controlled(Control_ref:Control ref)'
]

data = data.drop(columns=cols_to_drop, errors='ignore')

# Keep only numeric columns
data = data.select_dtypes(include=[np.number])

corr_matrix = data.corr(method='spearman')
corr_matrix2 = data.corr(method='pearson')

target = 'Penicillin concentration(P:g/L)'

pen_corr = (
    corr_matrix2[target]
    .drop(target)
    .sort_values(key=abs, ascending=False)
)

target = 'Offline Biomass concentratio(X_offline:X(g L^{-1}))'

bio_corr = (
    corr_matrix2[target]
    .drop(target)
    .sort_values(key=abs, ascending=False)
)

print(bio_corr.head(15))


```
